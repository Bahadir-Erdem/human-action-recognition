```python
# df kullanılarak her label için klasörlerin oluşturulması ve uygun resmin eklenmesi image1(calling) -> calling klasörüne gibi
train_folder_path = './images/train/'
test_folder_path = './images/test/'

for index, row in df.iterrows():
    filename = row['filename']
    label = row['label']

    if filename.startswith('Image_'):
        source_folder = train_folder_path
    else:
        source_folder = test_folder_path

    destination_folder = os.path.join(source_folder, label)

    os.makedirs(destination_folder, exist_ok=True)

    source_path = os.path.join(source_folder, filename)
    destination_path = os.path.join(destination_folder, filename)
    shutil.move(source_path, destination_path)
```
